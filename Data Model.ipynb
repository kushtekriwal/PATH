```python
import pandas as pd
import numpy as np
```

```python
dhs = pd.read_csv("C:/Users/kusht/Downloads/DHS_Indicator_Data",encoding = 'unicode_escape')
```

```python
enap = pd.read_excel("C:/Users/kusht/Downloads/NEML and LMIS 2019 Every Newborn Tracking Tool Data .xlsx", header=1)
```

```python
enap = enap.iloc[:90]
```

```python
target = pd.DataFrame(columns=['Data Source', 'Country', 'Year', 'Characteristic Category', 'Indicator Name', 'Indicator Value'])
target['Country'] = dhs['CountryName']
target['Year'] = dhs['SurveyYear']
target['Characteristic Category'] = dhs['CharacteristicCategory']
#target['Characteristic Label'] = dhs['CharacteristicLabel']
target['Indicator Name'] = dhs['Indicator']
target['Indicator Value'] = dhs['Value']
target['Data Source'] = 'DHS'
```

```python
enap2 = pd.melt(enap, id_vars=['Countries'], var_name='Indicator Name', value_name='Indicator Value (Text)')
enap2['Indicator Value (Numeric)'] = np.nan
```

```python
for idx, i in enumerate(enap2['Indicator Name']):
    arr = i.split("_")
    ind = ""
    if arr[1] == 'NEML':
        ind = "Is " + arr[2] + " included in the National Essential Medicines List (NEML)?"
    else:
        ind = "Does the country have Logistics Mangament Information System (LMIS) for " + arr[2] + "?"
    enap2['Indicator Name'].loc[idx] = ind
```

```python
for idx, i in enumerate(enap2["Indicator Value (Text)"]):
    if i == "Yes":
        enap2["Indicator Value (Numeric)"].loc[idx] = 1
    elif i == "No":
        enap2["Indicator Value (Numeric)"].loc[idx] = 0
    elif i == "In process":
        enap2["Indicator Value (Numeric)"].loc[idx] = np.nan
    else:
        enap2["Indicator Value (Numeric)"].loc[idx] = np.nan
        enap2["Indicator Value (Text)"].loc[idx] = "Unknown"
```
