```python
spa = pd.read_excel("C:/Users/kusht/Downloads/SPA_KenyaTanzaniaMalawi_data.xlsx")
```

```python
spa['Indicator Name'].iloc[55:] = spa['Indicator Name (For Indicator GapMap)'].iloc[55:]
```

```python
spa['Data Source'] = 'DHS Service Provision Assesment (SPA)'
spa['Year'] = 0
```

```python
for idx, i in enumerate(spa['Indicator Value (Numeric)']):
    if i == 'Not measured':
        spa['Indicator Value (Numeric)'].loc[idx] = np.nan
```
