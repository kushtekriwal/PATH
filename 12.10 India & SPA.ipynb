```python
spa_clean['Characteristic Category'] = 'Total'
spa_clean['Characteristic Label'] = 'Total'
spa_clean['Country'] = spa['Country']
spa_clean['Data Source'] = 'DHS Service Provision Assesment (SPA)'
spa_clean['Indicator Name'] = spa['Indicator Name (match)']
spa_clean['Indicator Value (Numeric)'] = spa['SPA']
spa_clean['Indicator Value (Text)'] = np.nan
spa_clean['Year'] = np.nan
```

```python
for idx, i in enumerate(spa_clean['Indicator Value (Numeric)']):
    if i == 'Not measured':
        spa_clean['Indicator Value (Numeric)'].loc[idx] = np.nan
        spa_clean['Indicator Value (Text)'].loc[idx] = 'Not measured'
    elif i == 'Not in report':
        spa_clean['Indicator Value (Numeric)'].loc[idx] = np.nan
        spa_clean['Indicator Value (Text)'].loc[idx] = 'Not in report'
```
