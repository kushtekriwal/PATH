```python
import pandas as pd
import numpy as np
```

```python
hmis = pd.read_excel("C:/Users/kusht/Downloads/RMNCH HMIS data.xls", header=1)
```

```python
hmis
```

```python
hmis2 = pd.melt(hmis, id_vars=['orgunitlevel1', 'periodid', 'organisationunitname'], var_name='Indicator Name', value_name='Indicator Value (Numeric)')
```

```python
hmis2
```

```python
hmis2 = hmis2.rename(columns={'orgunitlevel1':'Country', 'periodid':'Year', 'organisationunitname':'Characteristic Label'})
hmis2['Data Source'] = 'HMIS'
hmis2['Characteristic Category'] = 'Region'
hmis2['Indicator Value (Text)'] = np.nan
```

```python
hmis2
```

```python
for idx, i in enumerate(hmis2['Characteristic Label']):
    if i == 'Kenya':
        hmis2['Characteristic Label'].loc[idx] = 'Total'
        hmis2['Characteristic Category'].loc[idx] = 'Total'
```

```python
hmis2
```

```python
model = pd.read_csv("C:/Users/kusht/Downloads/Final_11-19_v2.csv")
```

```python
model = model.append(hmis2, ignore_index=True)
```

```python
cols = model.columns.tolist()
cols
```

```python
cols = [cols[3]] + [cols[2]] + [cols[7]] + [cols[0]] + [cols[1]] + [cols[4]] + [cols[5]] + [cols[6]]
cols
```

```python
model = model[cols]
model
```
