```python
import pandas as pd
import numpy as np
```

```python
gl = pd.read_csv("C:/Users/kusht/Downloads/1210/Indicator Metadata All Countries_v3_12.9.2019.csv", encoding='latin_1')
```

```python
bf_sara = pd.read_excel("C:/Users/kusht/Downloads/sara/BurkinaFaso_SARA_data_2018.xlsx")
```

```python
bf_sara = bf_sara.rename(columns={'Source':'Data Source','Indicator Name (match)':'Indicator Name', 'Indicator Value':'Indicator Value (Numeric)'})
bf_sara = bf_sara.drop(['Asset', 'Milestone', 'Indicator Definition', 'notes'], axis=1)
bf_sara['Indicator Value (Text)'] = np.nan
bf_sara['Characteristic Category'] = 'Total'
bf_sara['Characteristic Label'] = 'Total'
```

```python
for idx, i in enumerate(bf_sara['Indicator Value (Numeric)']):
    if i == 'Not in report':
        bf_sara['Indicator Value (Numeric)'].loc[idx] = np.nan
        bf_sara['Indicator Value (Text)'].loc[idx] = 'Not in report'
```

```python
bf_sara
```

```python
bf_sara_join = pd.merge(gl, bf_sara, left_on=['Indicator Name (matched with data source)','Country'], right_on=['Indicator Name', 'Country'], how='right')
```

```python
bf_sara_join.columns.tolist()
bf_sara_join = bf_sara_join.drop(['Indicator Name (matched with data source)',
 'Indicator Source (linked to IS sheet)',
 'Indicator Value (Yes/No/NA). Asset included in policy/guidelines? Only applies to country-specific data (not "All Countries")'],axis=1)
```

```python
bf_sara_join = bf_sara_join.drop_duplicates(keep='first')
```

```python
bf_sara_join = bf_sara_join.iloc[:33]
```

```python
from IPython.display import FileLink, FileLinks

bf_sara_join.to_csv('C:/Users/kusht/Downloads/sara/BF_SARA_Processed.csv', index=False)
```
